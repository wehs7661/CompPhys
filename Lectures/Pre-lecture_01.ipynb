The exponential function can be calculated from:
$$e^x=\sum_{n=0}^\infty \frac{1}{n!}x^n$$

```python
import math

x = 1
exp_approx = 0.0
for n in range(0, 1001, 1):
    exp_approx += (x**n)/math.factorial(n)

print(exp_approx)
print(math.exp(x))
```

```python
import math
import numpy as np

x = 1.0
exp_approx = 0.0
for n in np.arange(0, 101, 1):
    exp_approx += (x**n)/math.factorial(n) 

print(exp_approx)
print(np.exp(x))
```
